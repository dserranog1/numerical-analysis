```python
from fractions import Fraction
from math import factorial
```

```python

# Define the coefficients as Fractions for exactness
a = [Fraction(0), Fraction(0), Fraction(-24), Fraction(24)]
b = [
    Fraction(1),
    Fraction(-5),
    Fraction(19),
    Fraction(9),
]

def d_j(j):
    """
    Computes d_j = sum_{i=0 to 5} [ (i^j / j!) * a_i - (i^(j-1) / (j-1)!) * b_i ],
    using Fraction for exact arithmetic.
    """
    total = Fraction(0)
    for i in range(4):
        # First term: (i^j / j!) * a_i
        print("para i=", i)
        term1 = Fraction(i**j, factorial(j)) * a[i]
        
        # Second term: (i^(j-1)/(j-1)!) * b_i, but skip if j=0
        if j == 0:
            term2 = Fraction(0)
        else:
            term2 = Fraction(i**(j-1), factorial(j-1)) * b[i]
        print(f"term1 = {term1}")
        print(f"term2 = {term2}")
        print("**")
        
        total += term1 - term2
    return total

for j_val in range(6):
    print(f"d{j_val} = ")
    val = d_j(j_val)
    print(f"d_{j_val} = {val}\n\n\n")
```
