# Order for Adams-Bashforth Method

```python
from fractions import Fraction
from math import factorial

# Define the coefficients as Fractions for exactness
a = [Fraction(0), Fraction(0), Fraction(0), Fraction(0), Fraction(-720), Fraction(720)]
b = [
    Fraction(251),
    Fraction(-1274),
    Fraction(2616),
    Fraction(-2774),
    Fraction(1901),
    Fraction(0),
]

def d_j(j):
    """
    Computes d_j = sum_{i=0 to 5} [ (i^j / j!) * a_i - (i^(j-1) / (j-1)!) * b_i ],
    using Fraction for exact arithmetic.
    """
    total = Fraction(0)
    for i in range(6):
        # First term: (i^j / j!) * a_i
        print("para i=", i)
        term1 = Fraction(i**j, factorial(j)) * a[i]
        
        # Second term: (i^(j-1)/(j-1)!) * b_i, but skip if j=0
        if j == 0:
            term2 = Fraction(0)
        else:
            term2 = Fraction(i**(j-1), factorial(j-1)) * b[i]
        print(f"term1 = {term1}")
        print(f"term2 = {term2}")
        print("**")
        
        total += term1 - term2
    return total

for j_val in range(7):
    print(f"d{j_val} = ")
    val = d_j(j_val)
    print(f"d_{j_val} = {val}\n\n\n")

```

### Como d1=d2=d3=d4=d5=0!=d6, concluimos que es de orden 5
